```python
# ---------------- Project 15: Stack Questions ----------------

# Function to display stack nicely
def show_stack(title, stack):
    print(f"{title}: {stack}")

# ---------------- Practical (Rwanda) 1 ----------------
momo_stack = []
momo_stack.append("Enter Code")
momo_stack.append("Enter PIN")
momo_stack.append("Confirm")

show_stack("MoMo Stack (before pop)", momo_stack)

# Pop last step
momo_stack.pop()
show_stack("MoMo Stack (after pop)", momo_stack)


# ---------------- Practical (Rwanda) 2 ----------------
ur_stack = []
ur_stack.append("Attend Class")
ur_stack.append("Write Assignment")
ur_stack.append("Sit Exam")

show_stack("UR Stack (before undo)", ur_stack)

# Undo two
ur_stack.pop()
ur_stack.pop()
show_stack("UR Stack (after undo)", ur_stack)


# ---------------- Challenge ----------------
challenge_stack = []
challenge_stack.append("X")
challenge_stack.append("Y")
challenge_stack.append("Z")

show_stack("Challenge (before pops)", challenge_stack)

# Pop twice
challenge_stack.pop()
challenge_stack.pop()

# Push W
challenge_stack.append("W")
show_stack("Challenge (final)", challenge_stack)

# Show top
print("Top element now:", challenge_stack[-1])


# ---------------- Reflection ----------------
print("\nReflection:")
print("A stack uses LIFO (Last In, First Out).")
print("This means the last person who comes is served first.")
print("For serving people in order (e.g., in a bank or MoMo agent),")
print("a queue (FIFO: First In, First Out) is more fair than a stack.")
```

```python
# Task: Push ["Enter Code", "Enter PIN", "Confirm"], then pop last step.

# Code (Python):

# MoMo stack example
stack = []
stack.append("Enter Code")
stack.append("Enter PIN")
stack.append("Confirm")

print("Stack after pushes:", stack)

# Pop last step
stack.pop()
print("Stack after pop:", stack)

# Output (screenshot style):

# Stack after pushes: ['Enter Code', 'Enter PIN', 'Confirm']
# Stack after pop: ['Enter Code', 'Enter PIN']


# ---

# 2. Practical (Rwanda): UR tasks

# Task: Push ["Attend Class", "Write Assignment", "Sit Exam"], then undo (pop) two.

# Code:

# UR student stack example
stack = []
stack.append("Attend Class")
stack.append("Write Assignment")
stack.append("Sit Exam")

print("Stack after pushes:", stack)

# Undo (pop) two
stack.pop()
stack.pop()
print("Stack after undo two:", stack)

# Output:

# Stack after pushes: ['Attend Class', 'Write Assignment', 'Sit Exam']
# Stack after undo two: ['Attend Class']


# ---

# 3. Challenge Question

# Task: Push ["X", "Y", "Z"], pop twice, then push "W". Find the top.

# Algorithm (Step-by-step):

# 1. Start with empty stack.


# 2. Push X - [X]


# 3. Push Y - [X, Y]


# 4. Push Z - [X, Y, Z]


# 5. Pop twice - [X]


# 6. Push W - [X, W]


# 7. Top = "W" ✅



# Code:

stack = []
stack.extend(["X", "Y", "Z"])
stack.pop()
stack.pop()
stack.append("W")
print("Top element is:", stack[-1])

# Output:

# Top element is: W


# ---

# 4. Reflection (Theory – no code)

# A stack uses Last In, First Out (LIFO) order.
# This is not good for serving people in real life (e.g., bank, hospital, MoMo agent) because the last person gets served first while those who came earlier wait unfairly.


# ---

# Part B: Queue Questions

# 1. Practical (Rwanda): RRA office

# Task: 4 people join queue. After 2 are served, who is front?

# Code:

from collections import deque

queue = deque(["P1", "P2", "P3", "P4"])
print("Queue:", queue)

# Serve two
queue.popleft()
queue.popleft()

print("Front person now:", queue[0])

# Output:

# Queue: deque(['P1', 'P2', 'P3', 'P4'])
# Front person now: P3


# ---

# 2. Practical (Rwanda): Nyabugogo buses

# Task: 3 buses enqueue. After 1 departs, which bus is next?

# Code:

queue = deque(["Bus1", "Bus2", "Bus3"])
print("Queue:", queue)

# First bus departs
queue.popleft()

print("Next bus:", queue[0])

# Output:

# Queue: deque(['Bus1', 'Bus2', 'Bus3'])
# Next bus: Bus2


# ---

# 3. Challenge Question (Clinic system)

# Algorithm:

# 1. Patients arrive and join a queue.


# 2. Each patient is served in order of arrival.


# 3. If a stack was used instead, last patient would be treated first.



# Problem:
# This causes unfairness – sick patients who came earlier might be ignored, which is dangerous in a clinic.


# ---

# 4. Reflection (Theory – no code)

# Queue (FIFO) = First In, First Out - fairness, order of arrival is respected.

# Stack (LIFO) = Last In, First Out - unfair in services, only good for temporary tasks (like undo, redo, MoMo steps).
```

```python
# -------------------------------
# Queue Questions in Python
# -------------------------------

from collections import deque

# ---------- Practical 1: RRA Office ----------
print("Practical 1: RRA Office")

# Queue of people
rra_queue = ["Alice", "Bob", "Celine", "David"]

# Serve 2 people
rra_queue.pop(0)  # Alice served
rra_queue.pop(0)  # Bob served

# Front of the queue
print("Front of the queue:", rra_queue[0])
print("-"*40)

# ---------- Practical 2: Nyabugogo Buses ----------
print("Practical 2: Nyabugogo Buses")

# Queue of buses
bus_queue = ["Bus A", "Bus B", "Bus C"]

# First bus departs
bus_queue.pop(0)

# Next bus
print("Next bus at front:", bus_queue[0])
print("-"*40)

# ---------- Challenge: Clinic Queue ----------
print("Challenge: Clinic Queue")

# Using deque for efficient queue operations
clinic_queue = deque()

# Patients arrive
clinic_queue.append("Patient 1")
clinic_queue.append("Patient 2")
clinic_queue.append("Patient 3")

# Serve first patient
served_patient = clinic_queue.popleft()
print("Served patient:", served_patient)
print("Next patient in line:", clinic_queue[0])
print("-"*40)

# ---------- Fairness Problem ----------
print("Fairness Problem:")

# If a stack (LIFO) is used instead
stack_queue = []

# Patients arrive
stack_queue.append("Patient 1")
stack_queue.append("Patient 2")
stack_queue.append("Patient 3")

# Serve patient from stack (LIFO)
served_lifo = stack_queue.pop()
print("Served patient using stack (LIFO):", served_lifo)
print("Next patient in line if stack used:", stack_queue[-1] if stack_queue else "None")
print("\nObservation: Last arrived patient served first, early arrivals wait longer (unfair).")
print("-"*40)

# ---------- Reflection ----------
print("Reflection:")
print("FIFO ensures first come, first served, representing fairness.")
print("LIFO can delay early arrivals, which is unfair for real-life queues like clinics or offices.")
```
